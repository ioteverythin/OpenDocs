# Data Analysis Report

This notebook demonstrates the analysis of sales data for Q4 2025.

## Overview

We analyze trends, patterns, and key metrics from the quarterly dataset.

```python
import numpy as np

# Generate sample data
months = ['Oct', 'Nov', 'Dec']
revenue = [28432, 27891, 32306]
units = [320, 285, 410]
```

```python
# Summary statistics
for m, r in zip(months, revenue):
    print(f'  {m}: ${r:,}')
```

## Key Findings

- **Revenue** averaged $29,543 per period
- **Unit sales** ranged from 285 to 410
- December showed the highest performance

### Recommendations

1. Increase marketing spend in October
2. Optimize inventory for December peak
3. Review pricing strategy for November

```python
total_rev = sum(revenue)
total_units = sum(units)
print(f'Total Revenue: ${total_rev:,}')
print(f'Total Units: {total_units:,}')
print(f'Average Revenue/Unit: ${total_rev/total_units:.2f}')
```
