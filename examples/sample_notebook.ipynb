# Data Analysis Report

This notebook demonstrates the analysis of sales data for Q4 2025.

## Overview

We analyze trends, patterns, and key metrics from the quarterly dataset.

```python
import pandas as pd
import numpy as np

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'month': ['Oct', 'Nov', 'Dec'] * 10,
    'revenue': np.random.uniform(10000, 50000, 30),
    'units': np.random.randint(100, 500, 30)
})
```

```python
# Summary statistics
data[['revenue', 'units']].describe()
```

## Key Findings

- **Revenue** averaged $29,543 per period
- **Unit sales** ranged from 102 to 498
- December showed the highest variance

### Recommendations

1. Increase marketing spend in October
2. Optimize inventory for December peak
3. Review pricing strategy for November dip

```python
# Monthly breakdown
monthly = data.groupby('month')['revenue'].mean()
print('Monthly Revenue Summary:')
for month, rev in monthly.items():
    print(f'  {month}: ${rev:,.2f}')
```
